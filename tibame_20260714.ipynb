- 柯拉茲猜想

- 對於一個數(n) 奇數: 3*n+1 偶數: n//2

- 猜想: 任何正整數都會收斂到1(16->8->4->2->1)?

- 想: 奇數(放大) 3*2+1 2+2+2+1 成為 偶數 / 2 -> 有可能是奇數, 也有可能是偶數

- 如果柯拉茲猜想是錯的, 那必須存在一個數字符合怎樣的格式? 奇數(x3)->偶數(/2) -> 奇數(本身1.5倍)
- 為什麼柯拉茲猜想很可能是對的, 1 < 2 < 4 < 8 < 16 < 32........

- 你在某些數字會發現希望(上升到非常高), 但在某一刻還是碰到了這個 1 2 4 8 16 32 的序列 -> 然後瞬間掉落: 冰雹理論

- case1. 給你一個數字, 證明他會回到1
- case2. 給你一個數字, 告訴我他最高升到哪個數字


````python
# 語法複習:
# 迴圈SOP
# 1. for...in: 群
# 2. for i in range(次): 固定次數 i=0...次-1
# 3. while True+break: 不固定次數的迴圈
n = 7
while True:
    print(n)
    if n == 1:
        break
    if n % 2 == 1:
        n = n * 3 + 1
    else:
        n = n // 2
````

````python
# 給你一個數字 10: 1~10都會收斂
def is_collatz_converge(n):
    # 我如何確定不收斂(準備一個很大的次數, 如果他超過次數還沒收斂, False)
    count = 0
    while True:
        # print(n)
        if n == 1:
            break
        if n % 2 == 1:
            n = n * 3 + 1
        else:
            n = n // 2
        count = count + 1
        # 如果這麼多次還沒到1, 那我覺得這n很有可能真不收斂
        if count == 999999999999999:
            return False
    return True

is_collatz_converge(29)
````
