```python
# *
# **
# ***
# ****
left, right = "*", " "
layer = 4
for i in range(layer):
    lcount = i + 1
    rcount = layer - i - 1
    line = left * lcount + right * rcount
    print(line)
```

```python
# line: 我要印幾行 total: 原本的問題是幾行
def func(line, total):
    # 不逃避
    if line == 0:
        pass
    # 逃避
    else:
        func(line-1, total)
        p = "*" * line + " " * (total - line)
        print(p)

func(5, 5)
```

```python
# 河內塔 有三個棍子, A棍子有n個圓盤-> C棍子, 幾步
# 一次只能移動一個, 大的一定要在小的下面
# n=3, step=7 n=4 step=15
def hanoi(n):
    # 不逃避
    if n == 1:
        return 1
    # 逃避
    else:
        # 我先找一個同學b幫我把上面三個般到棍子b
        step1 = hanoi(n-1)
        # 我把最下面的大盤子半到棍子c
        step2 = 1
        # 我再找一個同學c幫我把b的那三個班到棍子c
        step3 = hanoi(n-1)
        step = step1 + step2 + step3
        # step = 2 * hanoi(n-1) + 1
        return step
hanoi(3)
```

```python
def return_collatz_max_dp(n):
    # 不逃避
    if n == 1:
        return 1
    # 逃避
    else:
        if n % 2 == 0:
            # 16 -> 找出8的柯拉茲序列最大值
            other = return_collatz_max_dp(n // 2)
            ans = max(n, other)
        else:
            other = return_collatz_max_dp(3 * n + 1)
            ans = max(n, other)
        return ans

return_collatz_max_dp(7)
```

```python
import time

record = {}
def return_collatz_max_dp(n):
    # 不逃避
    if n == 1:
        return 1
    # 不逃避的額外case(DP)
    elif n in record:
        return record[n]
    # 逃避
    else:
        if n % 2 == 0:
            # 16 -> 找出8的柯拉茲序列最大值
            other = return_collatz_max_dp(n // 2)
            ans = max(n, other)
        else:
            other = return_collatz_max_dp(3 * n + 1)
            ans = max(n, other)
        # ! 開關
        record[n] = ans
        return ans

# 測一下時間
start_t = time.time()
result = {}
n_max = 10000000
for i in range(n_max):
    n = i + 1
    ans = return_collatz_max_dp(n)
    result[n] = ans
end_t = time.time()
dif_t = end_t - start_t
print("時間是:", dif_t)
# print("紀錄:", record)
# print("結果:", result)
```

```python
import matplotlib.pyplot as plt

x = []
y = []
for i in range(2, 10000000):
    x.append(i)
    y.append(result[i])

plt.plot(x, y)
plt.show()
```
