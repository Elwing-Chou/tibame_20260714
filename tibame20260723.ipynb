```python
# *
# **
# ***
# ****
left, right = "*", " "
layer = 4
for i in range(layer):
    lcount = i + 1
    rcount = layer - i - 1
    line = left * lcount + right * rcount
    print(line)
```

```python
# line: 我要印幾行 total: 原本的問題是幾行
def func(line, total):
    # 不逃避
    if line == 0:
        pass
    # 逃避
    else:
        func(line-1, total)
        p = "*" * line + " " * (total - line)
        print(p)

func(5, 5)
```
